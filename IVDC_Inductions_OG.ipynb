```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential, layers
from tensorflow.keras.layers import Dense, Conv2D, MaxPooling2D, Flatten, BatchNormalization, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.applications.resnet50 import preprocess_input
import pandas as pd

# Load your data
train_ds = keras.utils.image_dataset_from_directory(
    directory='/kaggle/input/tttttt/archive (1)/traffic_Data/DATA',
    labels='inferred',
    batch_size=32,
    image_size=(236, 236)
)

validation_ds = keras.utils.image_dataset_from_directory(
    directory='/kaggle/input/tttttt/archive (1)/traffic_Data/Test1',
    labels="inferred",
    batch_size=32,
    image_size=(236, 236)
)

# Normalize your data
def process(image, label):
    image = tf.cast(image / 255., tf.float32)
    return image, label

train_ds = train_ds.map(process)
validation_ds = validation_ds.map(process)

# One-hot encode the labels
num_classes = 58  # Assuming you have 58 classes
train_ds = train_ds.map(lambda x, y: (x, tf.one_hot(y, num_classes)))
validation_ds = validation_ds.map(lambda x, y: (x, tf.one_hot(y, num_classes)))

from tensorflow.keras.applications import EfficientNetB0,MobileNetV2
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Flatten, Dense, Dropout

# Load MobileNet model without including the top layer (fully connected layers)
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(236, 236, 3))

# Freeze the layers in the base MobileNet model
for layer in base_model.layers:
    layer.trainable = False

# Create your model architecture by adding layers on top of MobileNet
x = base_model.output
x = Flatten()(x)
x = Dense(128, activation='relu')(x)
x = Dropout(0.1)(x)
x = Dense(64, activation='relu')(x)
x = Dropout(0.1)(x)
output = Dense(58, activation='softmax')(x)

# Combine the base MobileNet model with the custom top layers
model = Model(inputs=base_model.input, outputs=output)


# Model definition
# model = Sequential([
#     Conv2D(32, kernel_size=(3, 3), padding='valid', activation='relu', input_shape=(236, 236, 3)),
#     BatchNormalization(),
#     MaxPooling2D(pool_size=(2, 2), strides=2, padding='valid'),

#     Conv2D(64, kernel_size=(3, 3), padding='valid', activation='relu'),
#     BatchNormalization(),
#     MaxPooling2D(pool_size=(2, 2), strides=2, padding='valid'),

#     Conv2D(128, kernel_size=(3, 3), padding='valid', activation='relu'),
#     BatchNormalization(),
#     MaxPooling2D(pool_size=(2, 2), strides=2, padding='valid'),

#     Flatten(),

#     Dense(128, activation='relu'),
#     Dropout(0.1),
#     Dense(64, activation='relu'),
#     Dropout(0.1),
#     Dense(num_classes, activation='softmax')
# ])

# # Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(train_ds, epochs=5)
```

```python
import numpy as np

# Predict labels for validation dataset
predictions = model.predict(validation_ds)
df = pd.read_csv('/kaggle/input/tttttt/archive (1)/labels.csv')
class_dict = dict(zip(df['ClassId'], df['Name']))
# Convert predicted probabilities to class labels
predicted_labels = np.argmax(predictions, axis=1)
labels = []
for i in predicted_labels:
    labels.append(class_dict[i])
```
